We are going to build a model that predicts if someone who seeks a loan might be a defaulter or a non-defaulter. We have several
independent variables like checking account balance, credit history, purpose, loan amount etc. For more details on the dataset, 
please see source at
https://archive.ics.uci.edu/ml/datasets/statlog+(german+credit+data)

Target variable is the person is the defaulter or not

# Import Libraries

```python
import pandas as pd
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
creditData = pd.read_csv("credit.csv")
creditData.head(10) #Several missing values
```

```python
creditData.shape
```

```python
creditData.describe()
```

```python
creditData.info()
#many columns are of type object i.e. strings. These need to be converted to ordinal type
```

Lets convert the columns with an 'object' datatype into categorical variables

```python
for feature in creditData.columns: # Loop through all columns in the dataframe
    if creditData[feature].dtype == 'object': #apply only for columns with categorical strings
        creditData[feature] = pd.Categorical(creditData[feature]) #Replace strings with an integer

creditData.head(10)
```

```python
creditData.info()
```

```python
print(creditData.checking_balance.value_counts())
print(creditData.credit_history.value_counts())
print(creditData.purpose.value_counts())
print(creditData.savings_balance.value_counts())
print(creditData.employment_duration.value_counts())
print(creditData.other_credit.value_counts())
print(creditData.housing.value_counts())
print(creditData.job.value_counts())
print(creditData.phone.value_counts())
print(creditData.dependents.value_counts())
```

```python
#converting categorical data


oneHotCols = ["purpose", "housing", "other_credit", "job"] #one hot coding only on these columns
```

```python

creditData = pd.get_dummies(creditData, columns=oneHotCols)
creditData.head(10)
```

```python
replaceStruct = {
                "checking_balance": {"< 0 DM": 1, "1 - 200 DM": 2, "> 200 DM": 3, "unknown": -1},
                "credit_history": {"critical": 1, "poor": 2, "good": 3, "very good": 4, "perfect": 5},
                "savings_balance": {"< 100 DM": 1, "100 - 500 DM": 2, "500 - 1000 DM": 3, "> 1000 DM": 4, "unknown": -1},
                "employment_duration": {"unemployed": 1, "< 1 year": 2, "1 - 4 years": 3, "4 - 7 years": 4, "> 7 years": 5},
                "phone": {"no": 1, "yes": 2},
                "default": {"no" : 0, "yes": 1}
}
creditData = creditData.replace(replaceStruct)
creditData.head(10)
```

```python
creditData.info()
```

# Split Data

```python
X = creditData.drop("default", axis=1)
y = creditData.pop("default")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

# Build Decision Tree Model

We will build model using the DecisionTreeClassifier function. Using default 'gini' criteria to split. Other option include 'entropy'

```python
dTree = DecisionTreeClassifier(criterion='gini', random_state=1)
dTree.fit(X_train, y_train)
```

Scoring our Decision Tree

```python
# Overfitting
print(dTree.score(X_train, y_train))
print(dTree.score(X_test, y_test))
```

Reducing OverFitting (Regularization)

```python

dTreeRegu = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=1)
dTreeRegu.fit(X_train, y_train)
print(dTreeRegu.score(X_train, y_train))
print(dTreeRegu.score(X_test, y_test))
```

```python
print(dTreeRegu.score(X_test, y_test))
y_predict = dTreeRegu.predict(X_test)

cm = metrics.confusion_matrix(y_test, y_predict, labels=[0,1])

df_cm = pd.DataFrame(cm, index = [i for i in ["No", "Yes"]], columns=[i for i in ["No", "Yes"]])

plt.figure(figsize=(7,5))
sns.heatmap(df_cm, annot=True, fmt='g')
```

# Ensemble Learning - Bagging

```python
from sklearn.ensemble import BaggingClassifier
```

```python
from sklearn.ensemble import BaggingClassifier

bgcl = BaggingClassifier(n_estimators=66, base_estimator=dTree, random_state=1)

bgcl = bgcl.fit(X_train, y_train)
```

```python
y_predict = bgcl.predict(X_test)

print(bgcl.score(X_train, y_train))
print(bgcl.score(X_test, y_test))

cm = metrics.confusion_matrix(y_test, y_predict, labels=[0,1])

df_cm = pd.DataFrame(cm, index = [i for i in ["No", "Yes"]], columns=[i for i in ["No", "Yes"]])

plt.figure(figsize=(7,5))
sns.heatmap(df_cm, annot=True, fmt='g')
```

XGboost

```python
from sklearn.ensemble import AdaBoostClassifier

abcl = AdaBoostClassifier(n_estimators=15, random_state=1)

abcl = abcl.fit(X_train, y_train)
```

# Ensemble Learning - GradientBoost

```python
from sklearn.ensemble import GradientBoostingClassifier

gbcl = GradientBoostingClassifier(n_estimators=76, random_state=1) # n_estimators means number of decision tree
                                                                                
gbcl = gbcl.fit(X_train, y_train) # fitting with training data
```

```python
y_predict = gbcl.predict(X_test)

print(gbcl.score(X_train, y_train))
print(gbcl.score(X_test, y_test))


cm=metrics.confusion_matrix(y_test, y_predict, labels=[0, 1])

df_cm = pd.DataFrame(cm, index = [i for i in ["No", "Yes"]], columns=[i for i in ["No", "Yes"]])

plt.figure(figsize=(7,5))

sns.heatmap(df_cm, annot=True, fmt='g')
```

# Ensemble RandomForest Classifier
